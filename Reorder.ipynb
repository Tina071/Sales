```python
import pandas as pd
import numpy as np
```

```python
aisle = pd.read_csv('aisles.csv')
department = pd.read_csv('departments.csv')
order_product = pd.read_csv('order_products.csv')
order = pd.read_csv('orders.csv')
product = pd.read_csv('products.csv')
```

```python
df1 = pd.merge(product, aisle, on='aisle_id', how='left')
```

```python
df2 = pd.merge(df1, department, on='department_id', how='left')
```

```python
df3 = pd.merge(order_product, df2, on='product_id', how='left')
```

```python
table_new = pd.merge(order, df3, on='order_id', how='left')
table_new
```

```python
table_new['order_rank'] = table_new.groupby(['user_id','product_id'])['order_number'].rank()
```

```python
def reorder(order_rank):
    if order_rank == 1:
        n = 0
    else: 
        n = 1
    return n
table_new['reordered'] = table_new['order_rank'].apply(reorder)
table_new.head(20)
```
