```python
import pandas as pd
import re
```

```python
product = pd.read_csv('products.csv')
product.head(50)
```

```python
regex = re.compile('[^\w\s]')
```

```python
product.product_name = product.product_name.str.replace(regex,'')
product.head(50)
```

```python
product.to_csv("product_clean.csv",index=False)
```
