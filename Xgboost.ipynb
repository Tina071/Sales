Deal with train dataset

```python
missing_total = train.isnull().sum().sort_values(ascending=False)
missing_total
```

```python
train.drop("Cabin", inplace=True, axis=1)
```

```python
train['Age'] = train.groupby(['Pclass', 'Sex'])['Age'].apply(lambda x: x.fillna(x.mean()))
train.head(20)
```

```python
train.drop("Name", inplace=True, axis=1)
```

```python
train.drop("Ticket", inplace=True, axis=1)
```

```python
train["Sex"] = train["Sex"].map({'male':1, 'female':0})
```

```python
embarked = pd.get_dummies(train["Embarked"], drop_first=True)
```

```python
train = pd.concat([train, embarked], axis=1)
```

```python
train.drop("Embarked", inplace=True, axis=1)
```

Random Forest Model

```python
X_train = train.drop("Survived", axis=1)
Y_train = train["Survived"]
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error

data_dmatrix = xgb.DMatrix(data=X_train,label=Y_train)
```

```python
xg_reg = xgb.XGBClassifier(objective ='reg:linear',
                           subsample  = 0.8,
                           colsample_bytree = 0.8, 
                           learning_rate = 0.1,
                           max_depth = 5, 
                           n_estimators = 10)
```

```python
xg_reg.fit(X_train,Y_train)
preds_train = xg_reg.predict(X_train)
```

```python
rmse = np.sqrt(mean_squared_error(Y_train, preds_train))
rmse
```

```python
params = {
          'eta': [0.03, 0.1, 0.2],
          'colsample_bytree': [0.5, 0.7, 0.9],
          'max_depth': range(3, 5, 10),
          'subsample': [0.5, 0.7, 0.9]
          }

grid_search = GridSearchCV(estimator = xg_reg, param_grid = params, scoring='roc_auc', cv=5)

grid_search.fit(X_train, Y_train)
grid_search.best_params_, grid_search.best_score_
```

```python
clf = xgb.train(params=grid_search.best_params_, dtrain=data_dmatrix, num_boost_round=10)
```

```python
xgb.plot_importance(clf)
plt.rcParams['figure.figsize'] = [5, 5]
plt.show()
```
